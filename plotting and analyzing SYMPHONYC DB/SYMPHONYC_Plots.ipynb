## Connecting my Database to the Jupyter Notebook and plot the data with the Matplot library!

## Import the required libraries:

```python
import pandas as pd
import pandas.io.sql as sqlio
import psycopg2 as ps
from IPython.core.display_functions import display
import matplotlib.pyplot as plt
plt.close("all")
import networkx as nx
```

## connect to the postgres:

```python
db = ps.connect(
                dbname = "SYMPHONYC DB",
                user = "postgres",
                password = "password",
                host = "localhost",
                port = "5432",
                options = '-c search_path="SYMPHONYC"'
)
```

## select * of artist table:

```python
sql_select_all_artist = """
                    SELECT *
                        FROM artist
"""
```

```python
artist_df = sqlio.read_sql_query(sql_select_all_artist, db)
artist_df_pie = artist_df.set_index('username')
```

```python
display(artist_df)
```

## plot all artists:

```python
artist_df.plot.bar(y='monthly_listeners', x='username', figsize=(7,7), color='DarkOrange')
plt.title("Chart of Artists vs. monthly_listeners")
plt.xlabel("Artists username")
plt.ylabel("number of monthly listeners")

artist_df.plot.scatter(y='monthly_listeners', x='username', figsize=(7,7), color='Yellow', s=80)
plt.title("Chart of Artists vs. monthly_listeners")
plt.xlabel("Artists username")
plt.ylabel("number of monthly listeners")

artist_df_pie.plot.pie(y='monthly_listeners', figsize=(12,12), autopct="%.2f", fontsize=16)
plt.title("Pie Chart of Artists vs. monthly_listeners in percent")
```

## select * of artist table ordered:

```python
sql_select_all_artist_ordered = """
                    SELECT *
                        FROM artist
                        ORDER BY monthly_listeners DESC
"""
artist_ordered_df = sqlio.read_sql_query(sql_select_all_artist_ordered, db)
artist_ordered_df_pie = artist_df.set_index('username')
display(artist_ordered_df)
```

```python
artist_ordered_df.plot.bar(y='monthly_listeners', x='username', figsize=(7,7), color='DarkOrange')
plt.title("Chart of Artists vs. monthly_listeners")
plt.xlabel("Artists username")
plt.ylabel("number of monthly listeners")

artist_ordered_df.plot.scatter(y='monthly_listeners', x='username', figsize=(7,7), color='Yellow', s=80)
plt.title("Chart of Artists vs. monthly_listeners")
plt.xlabel("Artists username")
plt.ylabel("number of monthly listeners")

artist_ordered_df_pie.plot.pie(y='monthly_listeners', figsize=(12,12), autopct="%.2f", fontsize=16)
plt.title("Pie Chart of Artists vs. monthly_listeners in percent")
```

## select * of albums table:

```python
sql_select_all_albums = """
                    SELECT *
                        FROM album
"""
albums_df = sqlio.read_sql_query(sql_select_all_albums, db)
#albums_df_pie = artist_df.set_index('username')
display(albums_df)
```

```python
albums_df.plot.scatter(y='release_date', x='artist_username', figsize=(18,7), color='red', s=50)
plt.title("Chart of Artists vs. their albums through the different years")
plt.xlabel("Artists username")
plt.ylabel("albums through the different years")
```

## select * of following_list table:

```python
sql_select_all_following_list = """
                    SELECT *
                        FROM following_list
"""
following_list_df = sqlio.read_sql_query(sql_select_all_following_list, db)
display(following_list_df)
```

```python
graph = nx.from_pandas_edgelist(following_list_df, 'follower', 'target', create_using=nx.DiGraph())
plt.figure(figsize=(18, 18))
nx.draw(graph, with_labels=True, arrows=True, node_size=1500, font_size=23, font_color="red", font_weight="bold", pos=nx.circular_layout(graph))
plt.title("Followers & Following graph network")
```

## song join album table:

```python
sql_song_join_album = """
                    SELECT *
	                    FROM song INNER JOIN album USING (album_id)
	                    WHERE artist_username = '32'
	                    ORDER BY album_id
"""
song_join_album_df = sqlio.read_sql_query(sql_song_join_album, db)
song_join_album_df.rename(columns={"number_of_times_played": "number of times song played", "number_of_likes": "number of likes its album has"}, inplace= True)
display(song_join_album_df)
```

```python
song_join_album_df.plot.bar(x='song_id', y=['number of times song played','number of likes its album has'] , figsize=(18,7))
plt.title("Chart of Artist(32) vs. his work stats")
plt.xlabel("Song ID")
plt.ylabel("Number of: ")
```

## Favorite Artists:

```python
sql_select_all_favorite_artist = """
                    SELECT person_username, artist_username
                        FROM favorite_artist
"""
favorite_artist_df = sqlio.read_sql_query(sql_select_all_favorite_artist, db)
display(favorite_artist_df)
```

```python
celeb_graph = nx.from_pandas_edgelist(favorite_artist_df, 'person_username', 'artist_username', create_using=nx.DiGraph())
plt.figure(figsize=(18, 18))
nx.draw(celeb_graph, with_labels=True, arrows=True, node_size=1500, font_size=23, font_color="red", font_weight="bold", pos=nx.circular_layout(celeb_graph))
plt.title("favorite artists graph network")
```
