## Connecting my Database to the Jupyter Notebook and plot the data with the Matplot library!

## Import the required libraries:

```python
import pandas as pd
import pandas.io.sql as sqlio
import psycopg2 as ps
from IPython.core.display_functions import display
import matplotlib.pyplot as plt
plt.close("all")
```

## connect to the postgres:

```python
db = ps.connect(
                dbname = "SYMPHONYC DB",
                user = "postgres",
                password = "password",
                host = "localhost",
                port = "5432",
                options = '-c search_path="SYMPHONYC"'
)
```

## select * of artist table:

```python
sql_select_all_artist = """
                    SELECT *
                        FROM artist
"""
```

```python
artist_df = sqlio.read_sql_query(sql_select_all_artist, db)
artist_df_pie = artist_df.set_index('username')
```

```python
display(artist_df)
```
